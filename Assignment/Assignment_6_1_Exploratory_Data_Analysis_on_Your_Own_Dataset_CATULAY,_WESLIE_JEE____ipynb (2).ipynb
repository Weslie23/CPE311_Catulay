```python
# Use numpy to get the average Kills and Assits

K = np.array(df['Kills'])
A = np.array(df['Deaths'])
AVK = int(np.average(K))
AVA = int(np.average(A))

print(f"This is the Average of Kills: {AVK}, and Average of Assists: {AVA}")
```

```python
# Use numpy to get the median Kills and Assits value.

KK = np.array(df['Kills'])
AA = np.array(df['Assists'])
M_K = int(np.median(KK))
M_A = int(np.median(AA))

print(f"This is the Median of Kills: {M_K}, and Median of Assists: {M_A}")
```

```python
# Use numpy to get the standard deviation of the skinthickness.

KKK = np.array(df['Kills'])
SVK= np.std(KKK)

print(f"This is the Standard Deviation of Kills: {SVK}")
```
